```python
import math
import numpy as np
Predicted =np.array([-0.5, 0.48, 3.1, 4.8, 5.5, 9.2])
actual = np.array([-0.45, 0.62, 3.25, 4.48, 5.22, 9.15])
n = len(actual)

ME = (1/n) * (np.sum(np.subtract(actual, Predicted)))
print('Mean Error: %.3f' % ME)
MAE = (1/n) * (np.sum(np.abs(np.subtract(actual, Predicted))))
print('Mean Absolute Error: %.3f' % MAE)


```

```python
mean_percent_error = 0
for i, j in zip(actual, Predicted):
    mean_percent_error += ((np.subtract(i,j))/i)*100

MPE = 1/n * (mean_percent_error)
print('Mean Percentage Error: %.3f' % MPE)

    
```

```python
mean_abs_percent_error = 0
for i, j in zip(actual, Predicted):
    mean_abs_percent_error += np.abs((np.subtract(i,j))/i)*100

MAPE = 1/n * (mean_abs_percent_error)
print('Mean Absolute Percentage Error: %.3f' % MAPE)
```

```python
RMSE = math.sqrt((1/n) * (np.sum(np.power(np.subtract(actual, Predicted), 2))))
print('Root Mean Square Error: %.3f' % RMSE)
```
