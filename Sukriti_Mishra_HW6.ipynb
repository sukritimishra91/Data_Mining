```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
array_rep = np.array(ratings)
```

```python
cosine_sim = {}
for i in range(0,49):
    cosine_sim[i+1] = cosine_similarity(array_rep[0,:].reshape(1,-1), array_rep[i+1,:].reshape(1,-1))[0][0]
    
```

```python
cosine_sim
```

### 3.	What pair of users do have the largest similarity (show pair and the corresponding cosine-similarity)? What pair of users do have the second largest similarity (show pair and the 

```python
key_val = max(cosine_sim, key=cosine_sim.get)
print('Cosine similarity for user 0 and user {} is highest and the value is {}'.format(key_val,round(cosine_sim[key_val],4)))
```

```python
n=2
result = {}
print(cosine_sim)
for key in sorted(cosine_sim, key=cosine_sim.get, reverse=True)[:n]:
    result.update({key: cosine_sim[key]})
print(result)
```

```python
count = 0
for key, value in result.items():
    if count ==0:
        print('Largest Cosine similarity for user 0 and user {} is highest and the value is {}'.format(key,result[key]))
    else:
        print('Second largest Cosine similarity for user 0 and user {} is highest and the value is {}'.format(key,result[key]))
    count+=1
```
